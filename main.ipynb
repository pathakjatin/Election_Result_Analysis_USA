```python
def clean(text):
    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', str(text))

    # Convert text to lowercase
    text = text.lower()

    # Replace anything other than alphabets a-z with a space
    text = re.sub('[^a-z]', ' ', text)

    # Split the text into single words
    text = text.split()

    # Initialize WordNetLemmatizer
    lm = WordNetLemmatizer()

    # Lemmatize words and remove stopwords
    text = [lm.lemmatize(word) for word in text if word not in set(
        stopwords.words('english'))]

    # Join the words back into a sentence
    text = ' '.join(word for word in text)

    return text
```

```python
def getpolarity(text):
    return TextBlob(text).sentiment.polarity

def getsubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

def getAnalysis(score):
    if score < 0:
        return 'negative'
    elif score == 0:
        return 'neutral'
    else:
        return 'positive'
```

```python
trump_tweets = data[data['candidate'] == 'trump']

# taking only U.S. country data
trump_tweets = trump_tweets.loc[trump_tweets.country == 'US']
trump_tweets = trump_tweets[['tweet']]
print(trump_tweets.head())
```

```python
# Define an optimized cleaning function
def clean(text):
    text = re.sub(r"http\S+", "", text)  # Remove URLs
    text = re.sub(r"@\w+", "", text)    # Remove mentions
    text = re.sub(r"#\w+", "", text)    # Remove hashtags
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)  # Remove special characters
    return text.strip().lower()  # Trim and convert to lowercase

# Apply the function using a vectorized operation
trump_tweets['cleantext'] = trump_tweets['tweet'].astype(str).apply(clean)

# Print the first few rows of the updated DataFrame
print(trump_tweets.head())
```

```python
def get_subjectivity(text):
    return TextBlob(text).sentiment.subjectivity

# Apply the function using vectorized operations
trump_tweets['subjectivity'] = trump_tweets['cleantext'].astype(str).apply(get_subjectivity)

# Print the first few rows of the updated DataFrame
print(trump_tweets.head())
```

```python
def get_polarity(text):
    return TextBlob(text).sentiment.polarity

# Apply the function to calculate polarity
trump_tweets['polarity'] = trump_tweets['cleantext'].astype(str).apply(get_polarity)

# Print the first few rows of the updated DataFrame
print(trump_tweets.head())
```

```python
# Define a function to classify the sentiment
def get_analysis(polarity):
    if polarity > 0:
        return 'Positive'
    elif polarity == 0:
        return 'Neutral'
    else:
        return 'Negative'

# Apply the function to classify the sentiment
trump_tweets['analysis'] = trump_tweets['polarity'].apply(get_analysis)

# Print the first few rows of the updated DataFrame
print(trump_tweets.head())
```

```python
# how much data is positive/negetive/neutral
plt.style.use('dark_background')  # Adding black theme

# Define colors for each bar
colors = ['orange', 'blue', 'red']

plt.figure(figsize=(7, 5))
(trump_tweets.analysis.value_counts(normalize=True) * 100).plot.bar(color=colors)
plt.ylabel("%age of tweets")
plt.title("Distribution of Sentiments towards Trump")
plt.show()
```
